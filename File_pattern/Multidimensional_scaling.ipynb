# Multidimensional scaling
Based on Manifold learning on handwritten digits: Locally Linear Embedding, Isomap 
https://scikit-learn.org/stable/auto_examples/manifold/plot_lle_digits.html


```python
from sklearn.datasets import load_digits

digits = load_digits(n_class=6)
X, y = digits.data, digits.target
n_samples, n_features = X.shape
n_neighbors = 30
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=10, ncols=10, figsize=(6, 6))
for idx, ax in enumerate(axs.ravel()):
    ax.imshow(X[idx].reshape((8, 8)), cmap=plt.cm.binary)
    ax.axis("off")
_ = fig.suptitle("A selection from the 64-dimensional digits dataset", fontsize=16)
```

```python
from sklearn.manifold import MDS
import matplotlib.pyplot as plt


# Define the MDS embedding with custom parameters
mds = MDS(n_components=2, n_init=1, max_iter=120, n_jobs=2, dissimilarity="euclidean")

# Fit the MDS embedding to your data
X_mds = mds.fit_transform(X)



# Create a scatter plot of the MDS results
plt.figure(figsize=(8, 6))
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y, cmap=plt.cm.nipy_spectral)
plt.title("MDS Embedding")
plt.colorbar()
plt.show()
```

```python
# Create a scatter plot of the MDS results with true labels as colors
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y, cmap=plt.cm.jet)

# Add true labels as text on the data points
for i in range(X_mds.shape[0]):
    plt.text(X_mds[i, 0], X_mds[i, 1], str(y[i]), fontsize=6, ha='center', va='center', color='white')

plt.title("MDS Embedding with True Labels")
plt.colorbar(scatter, label='True Labels')
plt.show()
```
