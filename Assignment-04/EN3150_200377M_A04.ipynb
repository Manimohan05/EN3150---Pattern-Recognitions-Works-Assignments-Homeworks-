Use the following mapping to map two-dimensional space to three-dimensional
space (feature space). This feature space set is known as projected set. Visualize
the projected set.


To map a two-dimensional space to a three-dimensional feature space using the mapping Φ, you can follow these steps:

Define the mapping function Φ(x).
Apply the mapping function to each data point in the original two-dimensional space to obtain the corresponding points in the three-dimensional space.
Visualize the projected set in the three-dimensional space.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the mapping function Φ(x)
def phi(x):
    x1, x2 = x
    phi_x1 = x1
    phi_x2 = x2
    phi_x3 = x1**2 + x2**2
    return phi_x1, phi_x2, phi_x3

# Generate some data points in the 2D space
np.random.seed(0)
num_points = 100
x2d = np.random.rand(num_points, 2)

# Apply the mapping to project the data to 3D space
x3d = np.array([phi(x) for x in x2d])

# Visualize the projected set in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x3d[:, 0], x3d[:, 1], x3d[:, 2])

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('Projected Set in 3D Space')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import 3D plotting functionality

# Generate the data in 2D space
np.random.seed(5)
X, y = make_circles(n_samples=500, factor=0.3, noise=0.1)

# Define the mapping function Φ
def phi_mapping(x):
    x1, x2 = x
    x3 = x1**2 + x2**2
    return np.array([x1, x2, x3])

# Apply the mapping to all data points
X_projected = np.apply_along_axis(phi_mapping, axis=1, arr=X)

# Create a 3D scatter plot to visualize the projected set
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot in 3D
ax.scatter(X_projected[:, 0], X_projected[:, 1], X_projected[:, 2], c=y, cmap=plt.cm.Paired)
ax.set_title('Projected Set in 3D Space')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

# Show the 3D plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data in 2D space with a different factor
np.random.seed(5)
X, y = make_circles(n_samples=500, factor=0.5, noise=0.1)

# Define the new mapping function x = (x1, x2) → (x1^2, x2^2, x1*x2)
def phi_mapping_new(x):
    x1, x2 = x
    return np.array([x1**2, x2**2, x1 * x2])

# Apply the new mapping to all data points
X_projected_new = np.apply_along_axis(phi_mapping_new, axis=1, arr=X)

# Create a 3D scatter plot to visualize the new projected set
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot in 3D
ax.scatter(X_projected_new[:, 0], X_projected_new[:, 1], X_projected_new[:, 2], c=y, cmap=plt.cm.Paired)
ax.set_title('Projected Set in 3D Space (New Mapping)')
ax.set_xlabel('Feature 1 (x1^2)')
ax.set_ylabel('Feature 2 (x2^2)')
ax.set_zlabel('Feature 3 (x1*x2)')

# Show the 3D plot with the new mapping
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_circles

# Generate the original dataset based on listing 1
np.random.seed(5)
X_orig, y_orig = make_circles(n_samples=500, factor=0.3, noise=0.1)

# Define the mapping function Φ(x)
def phi(x):
    x1, x2 = x
    phi_x1 = x1
    phi_x2 = x2
    phi_x3 = x1**2 + x2**2
    return phi_x1, phi_x2, phi_x3

# Apply the mapping to project the original data to 3D space
X_projected = np.array([phi(x) for x in X_orig])

# Split the data into training and testing sets for both datasets
X_orig_train, X_orig_test, y_orig_train, y_orig_test = train_test_split(X_orig, y_orig, test_size=0.2, random_state=42)
X_projected_train, X_projected_test, y_projected_train, y_projected_test = train_test_split(X_projected, y_orig, test_size=0.2, random_state=42)

# Create and train Linear SVC models on both datasets
clf_orig = SVC(kernel='linear')
clf_projected = SVC(kernel='linear')

clf_orig.fit(X_orig_train, y_orig_train)
clf_projected.fit(X_projected_train, y_projected_train)

# Make predictions on test data
y_orig_pred = clf_orig.predict(X_orig_test)
y_projected_pred = clf_projected.predict(X_projected_test)

# Calculate classification accuracies
accuracy_orig = accuracy_score(y_orig_test, y_orig_pred)
accuracy_projected = accuracy_score(y_projected_test, y_projected_pred)

print(f"Classification Accuracy on Original Dataset: {accuracy_orig:.2f}")
print(f"Classification Accuracy on Projected Dataset: {accuracy_projected:.2f}")
```
